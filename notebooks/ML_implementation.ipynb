```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('final_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df['customer_region'].unique()
```

```python
#Define Target Variable (High Demand)
#Threshold in terms of quantity_sold makes overfitted model with accuracy 1.0...So on checking with all viable columns...

threshold=df['total_revenue'].median()
df['High_Demand']=(df['total_revenue']>threshold)
df['High_Demand'].astype(int)
df.head()
```

```python
df=df.drop('order_date',axis=1)
```

```python
df=df.drop(['payment_method','review_count','rating'],axis=1)
```

```python
#Encoding Categorical Features...

df = pd.get_dummies(df, drop_first=False)
df.head()
```

```python
bool_cols = df.select_dtypes(include='bool').columns

df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.head()
```

```python
df.columns
```

```python
#Split Features and Target...

x = df.drop(['High_Demand','total_revenue'],axis=1)
y = df['High_Demand']
```

```python
x.shape
```

```python
y.shape
```

```python
#Train-Test Split...

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1000)
```

```python
print('X_train shape',x_train.shape)
print('X_test shape',x_test.shape)
print('Y_train shape',y_train.shape)
print('Y_test shape',y_test.shape)
```

```python
#Logistic Regression Implementation...

from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
logr.fit(x_train,y_train)
y_pred=logr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_pred))
print("Precision Score : ",precision_score(y_test,y_pred))
print("Recall Score : ",recall_score(y_test,y_pred))
print("F1 Score : ",f1_score(y_test,y_pred))
```

```python
#Naive Bayes Implementation..

from sklearn.naive_bayes import GaussianNB
```

```python
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
y_pred_nb = model_nb.predict(x_test)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_pred_nb))
print("Precision Score : ",precision_score(y_test,y_pred_nb))
print("Recall Score : ",recall_score(y_test,y_pred_nb))
print("F1 Score : ",f1_score(y_test,y_pred_nb))
```

```python
#KNN Implementation..

from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train, y_train)
y_pred_knn = model_knn.predict(x_test)
```

```python
print("Accuracy Score : ",accuracy_score(y_test,y_pred_knn))
print("Precision Score : ",precision_score(y_test,y_pred_knn))
print("Recall Score : ",recall_score(y_test,y_pred_knn))
print("F1 Score : ",f1_score(y_test,y_pred_knn))
```
